```python
df.shape
```

```python
# Outliers index by IQR
def outlier_index(df, col):
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        upper_limit= Q3+1.5*IQR
        lower_limit= Q1-1.5*IQR
        ls = df.index[(df[col] < lower_limit) | (df[col] > upper_limit)]
        return ls
```

```python
index_list=[]
for feature in ['price', 'hsize', 'property_age']:
    index_list.extend(outlier_index(df,feature))
```

```python
len(index_list)
```

```python
def remove(df, ls):
    ls=sorted(set(ls))
    df=df.drop(ls)
    return df
```

```python
df1= remove(df, index_list)
df1.shape
```

# Modeling

Splitting data:

```python
from sklearn.model_selection import train_test_split
X= df.drop('price', axis=1)
y= df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', X_test.shape)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def print_evaluate(true, predicted):  
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)

def regression(model,X_train, X_test, y_train, y_test): 
        # Training
        model.fit(X_train,y_train)
        train_preds = model.predict(X_train)
        print('Train set evaluation:\n________________________\n')
        print_evaluate(y_train, train_preds)
        
        print('\n====================================\n')
        
        # Tesing
        test_preds =model.predict(X_test)
        print('Test set evaluation:\n________________________\n')
        print_evaluate(y_test, test_preds)
        print('\n')
    
    
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
regression(model, X_train, X_test, y_train, y_test)
```

Let us try Ridge and Lasso regression (Regularization):

## Ridge Regression 

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=.1)
regression(model, X_train, X_test, y_train, y_test) 
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
model=Lasso(alpha=0.1)
regression(model, X_train, X_test, y_train, y_test)
```

We can see that the accuracy is getting significantly better using Ridge and Lasso regression. Now let us try Regression with  Random Forest, XGBoost, and Decision Tree Regressor, all of them can be used to predict our continuous variable 'price'.

## RandomForest

```python
%%time
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=0, n_estimators=200)
regression(model, X_train, X_test, y_train, y_test) 
```

## XGBoost Regressor

```python
from xgboost import XGBRegressor
model=XGBRegressor(n_estimators=400)
regression(model, X_train, X_test, y_train, y_test) 
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_features=100) 
regression(model, X_train, X_test, y_train, y_test) 
```

# PCA

Do you think Principal Components Analysis (PCA) could improve the model? Let us try:

```python
print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler=StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set
scaled_X_train=scaler.transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

But how many components are needed to describe the X_train?

```python
model_pca = PCA(n_components=177).fit(scaled_X_train)
plt.plot(np.cumsum(model_pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
model_pca = PCA(0.99)
# Fitting PCA on the training set only.
model_pca.fit(scaled_X_train)

new_X_train = model_pca.transform(scaled_X_train)
new_X_test  = model_pca.transform(scaled_X_test)
```

```python
models = {'Linear regression': LinearRegression(), 
          'Ridge': Ridge(alpha=0.1), 
          'Lasso': Lasso(alpha=0.1), 
          'Random Forest Regressor' : RandomForestRegressor(random_state=0, n_estimators=200),
          'XGBOOST Regressor': XGBRegressor(n_estimators=400), 'Decision tree Regressor':DecisionTreeRegressor()
}

         
for name, model in models.items():
    print('####################################################\n')
    print(name,':\n')
    print('........................\n')
    regression(model, new_X_train, new_X_test, y_train, y_test) 
```

```python
model_pca.n_components_
```

So, it is clear that PCA did not affect the accuracy and performance of model significantly.

# Linear Assumptions

We can observe that Ridge regression is the best model for our Data. Now let us check the linear assumptions:

```python
ridge_reg = Ridge(alpha=.1)
ridge_reg.fit(X_train,y_train)
y_pred = ridge_reg.predict(X_train)
```

```python
print(ridge_reg.score(X_train, y_train))
print(ridge_reg.score(X_test, y_test))
```

### 1- Mean of residuals

Mean of the residuals should be zero or very close to zero.

```python
residuals = y_train - y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```

### 2- Homoscedasticity- Constant Variance

```python
p = sns.scatterplot(y_pred,residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
p = sns.lineplot([0,175000],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

### 3- Residuals Normality

Residuals should be normal.

```python
plt.figure(figsize=(8,6))
p = sns.distplot(residuals,kde=True)
p = plt.title('Normality of error terms/residuals')
```

### 4- No autocorrelation of residuals\errors

Errors should not be related to each other, i.e we should not see any pattern in the errors.

```python
plt.figure(figsize=(10,5))
p = sns.lineplot(y_pred,residuals,marker='o',color='blue')
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')

p = sns.lineplot([0,175000],[0,0],color='red')
p = plt.title('Residuals vs fitted values plot for autocorrelation check')
```
